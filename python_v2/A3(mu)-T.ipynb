```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

```python
matplotlib.rc('text', usetex = True)
```

```python
mutf = pd.read_csv('muT123csv.csv',header = None)
```

```python
xutf = mutf[0]
mutf1 = mutf[1]
mutf2 = mutf[2]
mutf3 = mutf[3]
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman',size=12)
plt.rcParams['mathtext.fontset']='custom'
plt.rcParams['mathtext.rm']='Times New Roman'
plt.rcParams['font.family']='Times New Roman'
```

```python
plt.plot(xutf,mutf1,color='green', linestyle='solid',label='$T=0.15~GeV$')
plt.plot(xutf,mutf2,color='blue', linestyle='dashed',label='$T=0.3~GeV$')
plt.plot(xutf,mutf3,color='red', linestyle='dotted',label='$T=0.45~GeV$')
plt.xlabel("$\\mu_B$ [GeV]")
plt.ylabel("$A^3$ [GeV$^{-4}$fm$^{-3}$]")
plt.legend(["$T = 0.15$ GeV", "$T = 0.3$ GeV", "$T = 0.45$ GeV"], loc="lower right")
plt.grid(linestyle = '--', linewidth = 0.5)
plt.margins(x=0)
# plt.margins(y=0)
# text(10, 20, "function", fontsize=12)
# plt.show()
plt.savefig('A3_mu(T).pdf')
```
