```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

```python
matplotlib.rc('text', usetex = True)
```

```python
enf = pd.read_csv('energy.csv',header = None)
```

```python
xf = enf[0]
ef11 = enf[1]
ef21 = enf[2]
ef31 = enf[3]
ef12 = enf[4]
ef22 = enf[5]
ef32 = enf[6]
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman',size=12)
plt.rcParams['mathtext.fontset']='custom'
plt.rcParams['mathtext.rm']='Times New Roman'
plt.rcParams['font.family']='Times New Roman'
```

```python
plt.plot(xf,ef11,color='green', linestyle='solid')
plt.plot(xf,ef21,color='blue', linestyle='dashed')
plt.plot(xf,ef31,color='red', linestyle='dotted')
plt.plot(xf,ef12,color='green', linestyle='solid')
plt.plot(xf,ef22,color='blue', linestyle='dashed')
plt.plot(xf,ef32,color='red', linestyle='dotted')
plt.xlabel("$q / M$")
plt.ylabel("$E^{(r)}_{\\rm \\textit{\\textbf{p}}}$/M")
plt.legend(["\\textit{\\textbf{p}} = 0", "$p_{\\perp} = 0, p^3 = M$", "$p_{\\perp} = p^{3} = M$"], loc="upper left")
plt.grid(linestyle = '--', linewidth = 0.5)
plt.margins(x=0)
plt.margins(y=0)
# text(10, 20, "function", fontsize=12)
# plt.show()
plt.savefig('en.pdf')
```
