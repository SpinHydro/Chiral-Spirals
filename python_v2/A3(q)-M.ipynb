```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

```python
matplotlib.rc('text', usetex = True)
```

```python
mqf = pd.read_csv('qM345csv.csv',header = None)
```

```python
xqf = mqf[0]
mqf1 = mqf[1]
mqf2 = mqf[2]
mqf3 = mqf[3]
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman',size=12)
plt.rcParams['mathtext.fontset']='custom'
plt.rcParams['mathtext.rm']='Times New Roman'
plt.rcParams['font.family']='Times New Roman'
```

```python
plt.plot(xqf,mqf1,color='green', linestyle='solid',label='$M=0.3~GeV$')
plt.plot(xqf,mqf2,color='blue', linestyle='dashed',label='$M=0.4~GeV$')
plt.plot(xqf,mqf3,color='red', linestyle='dotted',label='$M=0.5~GeV$')
plt.xlabel("$q$ [GeV]")
plt.ylabel("$A^3$ [GeV$^{-4}$fm$^{-3}$]")
plt.legend(["$M = 0.3$ GeV", "$M = 0.4$ GeV", "$M = 0.5$ GeV"], loc="upper left")
plt.grid(linestyle = '--', linewidth = 0.5)
plt.margins(x=0)
plt.margins(y=0)
# text(10, 20, "function", fontsize=12)
# plt.show()
plt.savefig('A3_q(M).pdf')
```
